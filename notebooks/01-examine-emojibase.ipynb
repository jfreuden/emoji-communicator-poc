# 01: Examining the Emojibase Dataset
I know the emojibase dataset is a collection of dictionaries with emoji and tags, but I want to look at what words I have.

```python
from emoji import assets
import json

with open(assets.get_emojibase(), "r", encoding="utf-8") as file:
    data = json.load(file)

len(data)
```

```python
import pandas as pd

emoji_df = pd.DataFrame.from_dict(data)[['label', 'emoji', 'tags']].dropna()
emoji_df
```

```python
wordset = set()
wordcounts = dict()
wordemoji = dict()

for emoji, taglist in emoji_df[['emoji', 'tags']].itertuples(index=False, name=None):
    for tag in taglist:
        wordset.add(tag)
        wordcounts[tag] = wordcounts.get(tag, 0) + 1
        emojilist = wordemoji.get(tag, [])
        wordemoji[tag] = emojilist + [emoji]

word_df = pd.DataFrame.from_dict(wordcounts, orient='index').sort_values(by=0, ascending=False)
word_df = word_df.assign(wordemoji=wordemoji)
word_df
```
