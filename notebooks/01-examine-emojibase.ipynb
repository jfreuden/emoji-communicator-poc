# 01: Examining the Emojibase Dataset
I know the emojibase dataset is a collection of dictionaries with emoji and tags, but I want to look at what words I have.

```python
from emoji import assets
import json

with open(assets.get_emojibase(), "r", encoding="utf-8") as file:
    data = json.load(file)

len(data)
```
